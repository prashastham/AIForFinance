```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
start_date = '2010-10-22'
end_date = '2010-10-26'
dates = pd.date_range(start_date, end_date)
```

```python
df1 = pd.DataFrame(index=dates)
print(df1)
```

```python
dfINTC = pd.read_csv("data/INTC.csv", index_col='Date', parse_dates=True, usecols=['Date', 'Adj Close'], na_values='NaN')
dfAMD = pd.read_csv('data/AMD.csv', index_col='Date', parse_dates=True, usecols=['Date', 'Adj Close'], na_values='NaN')
```

```python
print(dfINTC)
```

```python
print(dfAMD)
```

```python
plt.rcParams["figure.figsize"] = [20, 8]
plt.rcParams["figure.autolayout"] = True
plt.plot(dfINTC['Adj Close'], '*-', color='red', markersize=1)
plt.plot(dfAMD['Adj Close'], '*-', color='green', markersize=1)
plt.show()
```
